```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

# Encode categorical variables.
encoder = OneHotEncoder(drop='first')
encoded_features = encoder.fit_transform(data[['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']])
```

```python
# Split data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(encoded_features, data['Rating'], test_size=0.2, random_state=42)

# Initialize and train the model.
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set.
y_pred = model.predict(X_test)

# Calculate RMSE and R-squared
rmse = mean_squared_error(y_test, y_pred, squared=False)
r_squared = r2_score(y_test, y_pred)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared: {r_squared}")
```

```python
# Test example.
genre = "Horror"
director = "Rahul Rawail"
actor1 = "Bobby Deol"
actor2 = "Aishwarya Rai Bachchan"
actor3 = "Shammi Kapoor"

# Create a dataframe with inputs.
user_input = pd.DataFrame({
    'Genre': [genre],
    'Director': [director],
    'Actor 1': [actor1],
    'Actor 2': [actor2],
    'Actor 3': [actor3]
})

# Preprocess user input using the encoder.
encoded_user_input = encoder.transform(user_input)

# Predict the rating using the trained model.
predicted_rating = model.predict(encoded_user_input)


print(f"Predicted Rating: {predicted_rating[0]}")
```
